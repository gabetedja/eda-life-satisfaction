```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
df=pd.read_csv("world-happiness-report.csv")
df.head()
```

# Column descriptions:
### Country name: 
Lists country name (the index since it's a unique value per year)
### Year: 
Lists year (2020 in this case by default removed from this future iterations of the data frame)
### Log GDP per capita: 
Natural logarithm of GDP per capita, log is taken because happiness does not increase at a linear rate with the GDP per capita
### Life ladder: 
A quantitative measure of subjective well being ranging from 0-10.
### Social support: 
Perception of availability of social support from peers or neighbors.
### Healthy life expectancy at birth: 
The amount of years a newborn is expected to live.
### Freedom to make life choices: 
Perceived freedom of choice in people's lives
### Generosity: 
Willingness to be charitable and generous to causes or peers.
### Perceptions of corruption: 
Public opinion on the presence ofcorruption in governments and institutions.
### Positive affect: 
Frequency of positive emotion (laughter, joy)
### Negative affect: 
Frequency of negative emotion (worry, sadness, anger)


```python
# FILTERS OUT BEFORE 2020
only_current_year_df = df[df.year == 2020].set_index('Country name').drop('year', axis=1)
only_current_year_df
```

```python
sorted_df = only_current_year_df.sort_values(by=['Life Ladder'], ascending=False)
sorted_df = sorted_df.dropna()
sorted_df
```

```python
gdp_df = sorted_df[["Life Ladder", "Log GDP per capita"]]
gdp_df
```

```python
plt.scatter(gdp_df['Log GDP per capita'], gdp_df['Life Ladder'])
plt.xlabel('Log GDP per capita')
plt.ylabel('Life Ladder')
plt.title('Log GDP per capita vs Life Ladder')
plt.show()
print("Correlation between the two variables: " + str(gdp_df['Log GDP per capita'].corr(gdp_df['Life Ladder'])))
```

```python
X = gdp_df[["Log GDP per capita"]]
y = gdp_df["Life Ladder"]
```

```python
model = LinearRegression()
model.fit(X, y) #model trained
```

```python
model.coef_ #find slope
```

```python
model.intercept_ #find intercept
```

```python
model.score(X, y) #find R^2
```

```python
preclean_2023_df = pd.read_csv("WHR2023.csv")
clean_2023_df = preclean_2023_df[['Country name', 'Logged GDP per capita']]
clean_2023_df = clean_2023_df.rename(columns={'Logged GDP per capita': 'Log GDP per capita'})
clean_2023_df = clean_2023_df.sort_values("Log GDP per capita", ascending=False).dropna()
clean_2023_df
```

```python
X_2023 = clean_2023_df[["Log GDP per capita"]]
predicted_life_ladder = model.predict(X_2023)
predicted_life_ladder

clean_2023_df['predicted scores'] = predicted_life_ladder
clean_2023_df
```

```python
actual_2023_df = pd.read_csv("WHR2023.csv")
actual_2023_df
```

```python
merged_df = pd.merge(clean_2023_df, actual_2023_df, on="Country name")
merged_clean_df = merged_df[["Country name", "Ladder score", "Log GDP per capita", "predicted scores"]]
merged_clean_df['error'] = np.abs(merged_clean_df["predicted scores"] - merged_clean_df["Ladder score"])
merged_clean_df
```

```python
mae = merged_clean_df['error'].mean()
mae
```
